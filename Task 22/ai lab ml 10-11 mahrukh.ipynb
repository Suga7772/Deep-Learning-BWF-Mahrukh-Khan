```python
import scipy as sc
import matplotlib.pyplot as plt
import numpy as np
```

```python
face = scipy.datasets.face()
plt.imshow(face)
```

```python
face_rotate = sc.ndimage.rotate(face, 135)
plt.imshow(face_rotate)
```

```python
flip_down = np.flipud(scipy.datasets.face())
plt.imshow(flip_down)
```
